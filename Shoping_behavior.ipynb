# About this file
The dataset explores how people shop based on their personal traits. It covers customer details like gender, age, income, and spending habits. With this information, analysts can understand customer predilection, shopping frequency, and the state between income levels and payment behavior.

```python
# import necessary liberary
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('Shoping.csv')
df.head()
```

## Presenting a data by Visulization

```python
df.head()
```

```python
# Showing the chat by  a age.

sns.histplot(x='Age', data=df)
plt.show()
```

```python
df.head()
```

```python
#chart bt gender and purchase Amount and payment methods

sns.countplot(x='Gender',data=df, hue= 'Payment Method')
plt.show()
```

```python
# making chart by item purchase 
item_counts = df['Item Purchased'].value_counts()

# Plot with percentages
plt.pie(
    item_counts.values,
    labels=item_counts.index,
    autopct='%1.1f%%'  # This formats the number to one decimal place
)

plt.title('Distribution of Items Purchased')
plt.show()
```

```python
#making graph by category 

sns.histplot(x='Category',data=df, hue='Gender')
plt.show()
```
